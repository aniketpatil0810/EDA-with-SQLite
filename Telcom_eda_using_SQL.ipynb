```python
# Step 1: Create the view
create_view = """
CREATE VIEW churnedcustomers AS
SELECT tenure, monthlycharges, contract
FROM customer
WHERE churn = 'Yes';
"""

# Step 2: Query from the view
query = "SELECT * FROM churnedcustomers;"
result = pd.read_sql(query, conn)
display(result)
```

```python
cursor = conn.cursor()

# Step 1: Drop view if it exists
cursor.execute("DROP VIEW IF EXISTS churn_rate_by_contract;")

# Step 2: Create the view
cursor.execute("""
CREATE VIEW churn_rate_by_contract AS
SELECT
    contract,
    COUNT(*) AS total_customers,
    SUM(CASE WHEN churn = 'Yes' THEN 1 ELSE 0 END) AS churned_customers,
    ROUND(
        CAST(SUM(CASE WHEN churn = 'Yes' THEN 1 ELSE 0 END) AS FLOAT) / COUNT(*) * 100,
        2
    ) AS churn_rate_percent
FROM customer
GROUP BY contract;
""")
conn.commit()

# Step 3: Query the view with pandas
query = "SELECT * FROM churn_rate_by_contract;"
result = pd.read_sql(query, conn)
display(result)
```

```python
# Step 1: Drop the view if it exists and create it
cursor = conn.cursor()
cursor.execute("DROP VIEW IF EXISTS revenue_per_internetservice;")
cursor.execute("""
CREATE VIEW revenue_per_internetservice AS
SELECT
    internetservice,
    SUM(totalcharges) AS total_revenue
FROM customer
GROUP BY internetservice;
""")
conn.commit()

# Step 2: Query the view with pandas
query = "SELECT * FROM revenue_per_internetservice;"
result = pd.read_sql(query, conn)
display(result)
```
